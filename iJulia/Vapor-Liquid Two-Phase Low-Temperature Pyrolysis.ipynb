```julia
using ReactionMechanismSimulator
using ReactionMechanismSimulator.Sundials
using ReactionMechanismSimulator.PythonPlot
```

# Load mechanism

```julia
input_file = "./PME/methyl_propyl_ether_solvation_kLAkH.rms"
phaseDict = readinput(input_file)
liqspcs = phaseDict["phase"]["Species"]
liqrxns = phaseDict["phase"]["Reactions"]
solvent = phaseDict["Solvents"][1]
gasspcs = liqspcs
gasrxns = []
liqspcnames = getfield.(liqspcs,:name)
gasspcnames = getfield.(gasspcs,:name);
```

# Create phases

```julia
gas = IdealGas(gasspcs,gasrxns;name="gas"); 
liq = IdealDiluteSolution(liqspcs,liqrxns,solvent;name="liq",diffusionlimited=true);
```

# Setting up domains, interfaces, and multi-domain reactor

```julia
# simulation conditions
T = 600.0
tf = 3600.0*24*12

# setting up liquid phase domain
Vliq = 1.0
PMEconc = 6154

initialconds = Dict("PME"=>PMEconc*Vliq,"T"=>T,"V"=>Vliq)
domainliq,y0liq,pliq = ConstantTVDomain(phase=liq,initialconds=initialconds,);

# setting up gas phase domain
Vgas = 1.0
P = 10*1e5
N2N = P*Vgas/R/T
initialconds = Dict("N2"=>N2N, "T"=>T,"P"=>P)
domaingas,y0gas,pgas = ConstantTPDomain(phase=gas,initialconds=initialconds);

# setting up gas phase outlet
# flow rate depends on the net evaporation/condensation
outletgas = VolumeMaintainingOutlet(domaingas)

# setting up mass transfer interface due to evaporation and condensation
# ignore the mass transfer of PME to prevent the liquid domain to dry up over time
ignoremasstransferspcnames = ["PME"]
vl,pinter = VaporLiquidMassTransferInternalInterfaceConstantT(domaingas,domainliq,ignoremasstransferspcnames);
```

# Setting up multi-domain reactor

```julia
domains = (domainliq,domaingas)
interfaces = [vl,outletgas]
react,y0,p = Reactor(domains,(y0liq,y0gas),(0.0,tf),interfaces,(pliq,pgas,pinter));
@time sol = solve(react.ode,react.recommendedsolver,abstol=1e-18,reltol=1e-6); #This take ~1 min
```

# Analyzing results

```julia
bsol = SystemSimulation(sol,domains, interfaces, p)
figure(dpi=300)
plotmolefractions(bsol.sims[1], sol.t[end-1], tol=1e-8, exclude=["N2","PME"])
```
